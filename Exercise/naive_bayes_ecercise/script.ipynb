```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB , MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_wine
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
data = load_wine(as_frame=True ).frame
```

```python
data
```

```python
data.info()
```

```python
x = data.drop(columns= ['target'])
y = data['target']

x_train ,x_test , y_train ,y_test = train_test_split(x,y,random_state=42,train_size=0.8 , stratify=y)

x_train.shape ,x_test.shape , y_train.shape ,y_test.shape
```

```python
g_model = GaussianNB()
g_model = g_model.fit(x_train , y_train)
```

```python
g_y_pred = g_model.predict(x_test)
```

```python
m_model = MultinomialNB()
m_model = m_model.fit(x_train , y_train)
```

```python
m_y_pred = m_model.predict(x_test)
```

```python
print('gausian accuracy_score : ' , accuracy_score(y_test , g_y_pred))
print('multinomial accuracy_score : ' , accuracy_score(y_test , m_y_pred))
```

```python
print(classification_report(y_test , g_y_pred))
print(classification_report(y_test , m_y_pred))
```
